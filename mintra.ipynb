```python
import os
import time 
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
product="anouk kurta"
#define the path to chrome webdriver executable

#Initialize the chrome Webdriver

driver=webdriver.Chrome()

#Navigate to URL
driver.get("https://www.myntra.com/kurta?rawQuery="+product)

myntra_text=driver.page_source

myntra_html=bs(myntra_text,"html.parser")
```

```python
myntra_text
```

```python
pclass=myntra_html.find_all('ul',{"class":"results-base"})
```

```python
pclass[0]
```

```python
for i in pclass:
    href=i.find_all("a",href=True)
    t=(href[0]['href'])
preoductLink="https://www.myntra.com/"+t
```

```python
driver.get(preoductLink)
praRes=driver.page_source
```

```python
proRes_html=bs(praRes,"html.parser")
```

```python
rat=proRes_html.find("a",{"class":"detailed-reviews-allReviews"})
t2=rat["href"]
review_link="https://www.myntra.com"+t2 
```

```python
driver.get(review_link)
```

```python
review=driver.page_source
review_html=bs(review,"html.parser")
```

```python
reviews=review_html.find_all("div",{"class":"detailed-reviews-userReviewsContainer"})
ratings=review_html.find_all("div",{"class":"user-review-main user-review-showRating"})
```

```python
ratings
```

```python

for i in reviews:
    customer_review=i.findAll("div",{"class":"user-review-reviewTextWrapper"})
    customer_ratings=i.findAll("div",{"user-review-starWrapper"})
```

```python
len(customer_ratings)
```

```python
for i in range(len(customer_review)):
    reviews=customer_review[i].text
    ratings=customer_ratings[i].text
    print(f"Ratings \n{ratings}\nRview\n{reviews}")
```
